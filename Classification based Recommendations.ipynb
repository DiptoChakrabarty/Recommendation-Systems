```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
bank_full =pd.read_csv("bank.csv")
```

```python
bank_full.head()
```

# Converting to Dummy Variables

```python
bank_updated=pd.get_dummies(bank_full,drop_first=True)
bank_updated.head()
```

```python
print("Updated : ",bank_updated.shape)
print("Original : ",bank_full.shape)

print("Updated columns: ",bank_updated.columns)
print("Original columns: ",bank_full.columns)
```

```python
X= bank_updated.drop(['deposit_yes'],axis=1)
y=bank_updated['deposit_yes']
```

# Regression Results

```python
logreg=LogisticRegression()
logreg.fit(X,y)
```
